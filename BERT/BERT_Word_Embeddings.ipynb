Loading Pre-trained BERT Model

```python
import torch
from transformers import BertTokenizer, BertModel
```

```python
# Load pre-trained model tokenizer (vocabulary)
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

Tokenization

```python
text = "BERT embeddings tutorial"
tokens = tokenizer.tokenize(text)
```

```python
token_ids = tokenizer.convert_tokens_to_ids(tokens)
```

Obtaining BERT Embeddings

```python
#Load pre-trained model
model = BertModel.from_pretrained('bert-base-uncased')
```

```python
# Convert the token IDs to a tensor
token_ids = torch.tensor(token_ids)

# Pass the token IDs through the BERT model
outputs = model(token_ids.unsqueeze(0))

# Extract embeddings from the hidden states
embeddings = outputs.last_hidden_state
```

```python
texts = ["BERT embeddings tutorial", "Hugging Face Transformers"]
# Tokenize the sentences and add padding to ensure the same length
max_length = max(len(tokenizer.tokenize(text)) for text in texts)
token_ids_batch = [tokenizer.encode(text, add_special_tokens=True, max_length=max_length, pad_to_max_length=True) for text in texts]
token_ids_batch = torch.tensor(token_ids_batch)

# Pass the batch through the BERT model
outputs_batch = model(token_ids_batch)

# Extract embeddings from the hidden states
embeddings_batch = outputs_batch.last_hidden_state
```
